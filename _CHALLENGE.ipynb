# TREND

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose as sm
#HERE THE START YEAR = 2000 AND END YEAR = 2020
j = np.arange(2000,2020)
trend = j * 2.75
plt.plot(j, trend)
```

# SEASONALITY

```python
s = 10 + np.sin(j) * 10
plt.plot(j, s)
```

# RESIDUAL

```python
r = np.random.normal(loc=0.0, scale=1, size=len(j))
plt.plot(j,r)
```

# ADDICTIVE TIME SERIES MODEL

```python
addictive_model = trend + s + r
plt.plot(j,addictive_model)
```

# MODEL EVALUATION USING TRAINING AND TESTING

```python
K = a.fillna(0)
print(K)
```

# DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

# FEATURE SELECTION

```python
g = ['JAHR','WERT','VORJAHRESWERT','VERAEND_VORMONAT_PROZENT','VERAEND_VORJAHRESMONAT_PROZENT','ZWOELF_MONATE_MITTELWERT']
X = K[g] 
y = a['MONAT']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
h = DecisionTreeClassifier()
q = h.fit(X_train,y_train)
y_pred = h.predict(X_test)
```

# MODEL EVALUATION

```python
print(metrics.classification_report(y_test, y_pred))
```
